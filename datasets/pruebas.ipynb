```python
# %pip install pandas
import pandas as pd

# Cargar ambos datasets
df_nano = pd.read_csv("dataset_NANO.csv")
df_onco = pd.read_csv("dataset_nano_ONCO.csv")

# Filas que están en dataset_NANO pero no en dataset_nano_ONCO (por Code)
# Un set es sin 'nano' y otro con 'nano' como palabra clave
codes_onco = set(df_onco["Code"])
df_diff = df_nano[~df_nano["Code"].isin(codes_onco)]

# Guardar las 54 filas eliminadas en un nuevo CSV
df_diff.to_csv("diferencias_filtradas.csv", index=False)

print(f"Filas guardadas en diferencias_filtradas.csv: {df_diff.shape[0]}")
```
